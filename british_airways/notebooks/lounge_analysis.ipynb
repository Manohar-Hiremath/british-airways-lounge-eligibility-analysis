```python
# Step 3: Grouping flights by key categories and calculating average lounge eligibility

# Group by Time of Day, Haul Type, and Destination Region
grouped_data = data.groupby(['TIME_OF_DAY', 'HAUL', 'ARRIVAL_REGION'])[
    ['TIER1_ELIGIBLE_PAX', 'TIER2_ELIGIBLE_PAX', 'TIER3_ELIGIBLE_PAX']
].mean().reset_index()

# Rename columns for clarity
grouped_data.rename(columns={
    'TIER1_ELIGIBLE_PAX': 'Avg_Tier1_Pax',
    'TIER2_ELIGIBLE_PAX': 'Avg_Tier2_Pax',
    'TIER3_ELIGIBLE_PAX': 'Avg_Tier3_Pax'
}, inplace=True)

# Display the grouped summary
display(grouped_data.head(15))
```

To better understand lounge demand patterns, we analyze how eligible passenger counts vary across different types of flights.
In this step, we group the flight data by three key operational dimensions:

Time of Day (Morning, Afternoon, Evening, Night)

Haul Type (Short-haul vs Long-haul)

Arrival Region (e.g., Europe, Asia, Middle East, Americas)

After grouping the data into these categories, we calculate the average number of eligible passengers for each British Airways lounge tier:

Tier 1 – Standard lounge access

Tier 2 – Premium lounge access

Tier 3 – Exclusive lounge access

The groupby() operation allows us to summarize how many passengers typically qualify for each lounge tier based on when and where the flight is operating.
We then rename the aggregated columns (e.g., Avg_Tier1_Pax) to make the output more readable and meaningful.

Finally, the first few rows of this grouped summary are displayed to give a clear view of how lounge eligibility varies across flight categories.
This step helps highlight important insights such as:

Which regions generally bring in more premium-tier passengers

How time-of-day influences lounge demand

Differences between short-haul and long-haul passenger eligibility

This summary forms the foundation for making data-driven decisions about lounge capacity planning and resource allocation.

```python
# Step 4: Create Lounge Eligibility Lookup Table (in percentages)

# Calculate total eligible passengers for normalization
grouped_data['Total_Pax'] = grouped_data['Avg_Tier1_Pax'] + grouped_data['Avg_Tier2_Pax'] + grouped_data['Avg_Tier3_Pax']

# Convert eligible passenger counts to percentages
grouped_data['Tier1_%'] = (grouped_data['Avg_Tier1_Pax'] / grouped_data['Total_Pax']) * 100
grouped_data['Tier2_%'] = (grouped_data['Avg_Tier2_Pax'] / grouped_data['Total_Pax']) * 100
grouped_data['Tier3_%'] = (grouped_data['Avg_Tier3_Pax'] / grouped_data['Total_Pax']) * 100

# Keep only relevant columns
lookup_table = grouped_data[['TIME_OF_DAY', 'HAUL', 'ARRIVAL_REGION', 'Tier1_%', 'Tier2_%', 'Tier3_%']]

# Round percentages for neatness
lookup_table = lookup_table.round(2)

# Display final lookup table
display(lookup_table.head(15))
```

This step converts the average number of eligible passengers in each tier into percentage values. The process begins by calculating the total number of eligible passengers for each group. Each tier’s share is then expressed as a percentage of this total, allowing the model to represent lounge eligibility in a standardised and scalable way.

Only the essential fields—time of day, haul type, arrival region, and the percentage values—are retained. The percentages are rounded for clarity, resulting in a concise lookup table that can be applied to estimate lounge demand across different flight schedules.

```python
# Step 5: Export the final lookup table to Excel

output_path = r"D:\BA_Lounge_Eligibility_Lookup.xlsx"
lookup_table.to_excel(output_path, index=False)

print("Lookup table successfully saved to Excel at:", output_path)
```

This step saves the finalized lookup table to an Excel file, allowing the processed lounge-eligibility percentages to be accessed outside the notebook. Exporting the data ensures the results are preserved, easy to share, and ready for future reporting or modeling tasks. This also creates a clean, structured output that can be directly used for presentations, dashboards, or operational planning without requiring the full dataset again.

```python
# ============================================================
# 📊 British Airways Lounge Eligibility Visualization (Final)
# ============================================================

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# -------------------------
# 1️⃣ Load the lookup table
# -------------------------
# If already loaded, skip this. Otherwise uncomment:
# lookup_table = pd.read_excel(r"D:\BA_Lounge_Eligibility_Lookup.xlsx")

print("✅ Data loaded successfully!")
print("Available columns:", lookup_table.columns.tolist())

# -------------------------
# 2️⃣ Create output directory
# -------------------------
output_dir = r"D:\BA_Visualization"
os.makedirs(output_dir, exist_ok=True)
chart_path = os.path.join(output_dir, "BA_Lounge_Tiers_BarChart.png")

# -------------------------
# 3️⃣ Visualization (Tier vs Eligible Passengers)
# -------------------------
plt.style.use("seaborn-v0_8-whitegrid")
plt.figure(figsize=(10, 6))

sns.barplot(
    data=lookup_table,
    x="Lounge Name",
    y="Eligible Passengers",
    palette="Blues_d"
)

# Titles and labels
plt.title("British Airways – Eligible Passengers per Lounge Tier", fontsize=16, fontweight="bold")
plt.xlabel("Lounge Name", fontsize=12)
plt.ylabel("Eligible Passengers", fontsize=12)
plt.xticks(rotation=30, ha="right", fontsize=10)

# Add value labels on top of bars
for index, row in enumerate(lookup_table["Eligible Passengers"]):
    plt.text(index, row + 0.5, f"{row:,}", ha='center', fontsize=9, color='black')

plt.tight_layout()
plt.savefig(chart_path, dpi=300, bbox_inches='tight')
plt.show()

print(f"\n✅ Professional bar chart created and saved at: {chart_path}")

import warnings
warnings.filterwarnings("ignore")
sns.barplot(data=lookup_table, x="Lounge Name", y="Eligible Passengers", hue="Lounge Name", palette="Blues_d", legend=False)

```
